# Histogram of an image

# In this code, we use the cv2.calcHist method to calculate the histogram of the image. This method takes the following arguments:

images: a list of the images to calculate the histogram for, in our case we only have one image.

channels: the channels to compute the histogram for, in this example we only have one channel (grayscale).

mask: an optional mask used to calculate the histogram for only a particular region of interest.

histSize: the number of bins to use in the histogram.

ranges: the range of pixel values to consider for the histogram.

After calculating the histogram, we use the matplotlib library to plot the histogram. The plt.plot method is used to plot the histogram values, and we set the x-axis limits to be between 0 and 256, since we have 256 bins in our histogram.

```python
#Histogram for grayscale
hist_gray = cv2.calcHist(gray_img, [0], None, [256], [0, 256])
plt.plot(hist_gray,color='k')
plt.show()
```

```python
#you can plot histogram same by splitting them
b,g,r=cv2.split(real_img)
```

```python
hist_red = cv2.calcHist([r],[0], None, [256], [0, 256])

# Plot the histogram
plt.plot(hist_red, color='r')
plt.xlim([0, 256])
plt.show()
```

```python
hist_green = cv2.calcHist([g], [0], None, [256], [0, 256])

# Plot the histogram
plt.plot(hist_green, color='g')
plt.xlim([0, 256])
plt.show()
```

```python
hist_blue = cv2.calcHist([b], [0], None, [256], [0, 256])

# Plot the histogram
plt.plot(hist_blue, color='b')
plt.xlim([0, 256])
plt.show()
```

```python
#Ploting them in one frame
plt.subplot(2,3,1)
plt.title('Original_image')
plt.imshow(real_img)

plt.subplot(2,3,2)
plt.title('Grayscale')
plt.imshow(gray_img,cmap='gray')

plt.subplot(2,3,3)
plt.title('Histogram_grayscale')
plt.plot(hist_gray,color='k')

plt.subplot(2,3,4)
plt.title('Histogram_red')
plt.plot(hist_red,color='r')

plt.subplot(2,3,5)
plt.title('Histogram_green')
plt.plot(hist_green,color='g')

plt.subplot(2,3,6)
plt.title('Histogram_blue')
plt.plot(hist_blue,color='b')

plt.show()
```

```python
#For Masked image
mask=np.zeros(gray_img.shape,np.uint8)
mask[300:900,100:900]=255
masked_img=cv2.bitwise_and(gray_img,gray_img,mask=mask)
hist=cv2.calcHist([masked_img],[0],mask,[256],[0,256])
plt.imshow(masked_img,cmap='gray')
plt.show()

plt.plot(hist)
plt.show()
```

# Edge Detector

```python
plt.subplot(1,2,1)
plt.title('original image')
plt.imshow(real_img)

plt.subplot(1,2,2)
plt.title('edged_detected')
plt.imshow(edge_real_img)

plt.show()
```

```python
#By Sobel
sobelx_img1=cv2.Sobel(src=real_img,ddepth=cv2.CV_64F,dx=1,dy=0,ksize=5)
sobely_img1=cv2.Sobel(src=real_img,ddepth=cv2.CV_64F,dx=0,dy=1,ksize=5)
sobelxy_img1=cv2.Sobel(src=real_img,ddepth=cv2.CV_64F,dx=1,dy=1,ksize=5)

cv2.imshow('img',sobelx_img1)
cv2.waitKey(0)

cv2.imshow('img',sobely_img1)
cv2.waitKey(0)

cv2.imshow('img',sobelxy_img1)
cv2.waitKey(0)
```

#  Fourier Transform to an image

```python
# Compute the discrete Fourier Transform of the image
fourier = cv2.dft(np.float32(gray_img), flags=cv2.DFT_COMPLEX_OUTPUT)

# Shift the zero-frequency component to the center of the spectrum
fourier_shift = np.fft.fftshift(fourier)

# calculate the magnitude of the Fourier Transform
magnitude = 20*np.log(cv2.magnitude(fourier_shift[:,:,0],fourier_shift[:,:,1]))

# Scale the magnitude for display
magnitude = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8UC1)

# Display the magnitude of the Fourier Transform
cv2.imshow('Fourier Transform', magnitude)
cv2.waitKey(0)
```
